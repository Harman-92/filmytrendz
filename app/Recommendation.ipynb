```python
#Read dataset
movies=pd.read_csv('movie_data.csv')
```

```python
#Get movie ids based on genre
rec_genre= pd.json_normalize(ts.Movies(id='27205').similar_movies()['results']).id.tolist()
rec_genre
```

```python
#Get movies from db based on tmdb_id
movies[movies.tmdb_id.isin(rec_genre)].head(5)
```

```python
#Get movie ids general
rec_gen= pd.json_normalize(ts.Movies(id='27205').recommendations()['results']).id.tolist()
rec_gen

#Similarly query db based on tmdb_id as in previous cell
```

```python
#Recommendations based on director
rec_dir=set(rec_gen+rec_genre)
rec_dir=list(rec_dir)
rec_dir
```
